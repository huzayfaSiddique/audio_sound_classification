```python
import os
audio_dataset_path='/content/drive/MyDrive/datasets/soundata/audio'
```

```python
def features_extractor(file):
  audio,sample_rate=librosa.load(file,res_type='kaiser_best')
  mfccs_features=librosa.feature.mfcc(y=audio,sr=sample_rate,n_mfcc=30)
  mfccs_features_scaled=np.mean(mfccs_features.T,axis=0)

  return mfccs_features_scaled
```

```python
import numpy as np
from tqdm import tqdm
extracted_features=[]
for i,j in tqdm(metadata.iterrows()):
  file_name=os.path.join('/content/drive/MyDrive/datasets/soundata/audio','fold'+str(j["fold"]),j['slice_file_name'])
  final_class_labels=j['class']
  extracted_data=features_extractor(file_name)
  extracted_features.append([extracted_data,final_class_labels])
```

```python
### converting extracted_features to Pandas dataframe
extracted_features_df=pd.DataFrame(extracted_features,columns=['feature','class'])
extracted_features_df.head()
```

```python
extracted_features_df.to_pickle('/content/drive/MyDrive/datasets/extracted_features_df.pkl')
```

```python
extracted_features_df=pd.read_pickle('/content/drive/MyDrive/datasets/extracted_features_df.pkl') # i have saved my model here in order
# to save the amount of training I have done before so every time i run my model i donot have to load it from scratch.
```

```python
#Split the dataset into independent and dependent features
X=np.array(extracted_features_df['feature'].tolist())
y=np.array(extracted_features_df['class'].tolist())
```

```python
X.shape
```

```python
y
```

```python
# one hot encoding vs label encoder and then to_categorical from tensorflow
# one hot encoding or .get_dummies() takes the class variables are do one hot encoding alphabetically while
# label encoder gives u a leverage that u can change and get a control over the order of the classes and u can give a specfic number to
#a class depending on the circumstance or your will
```

```python

from tensorflow.keras.utils import to_categorical
from sklearn.preprocessing import LabelEncoder
labelencoder=LabelEncoder()
y=to_categorical(labelencoder.fit_transform(y))
```

```python
y
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout,Activation,Flatten
from sklearn import metrics
from tensorflow.keras.optimizers import Adam
```

```python
num_lables=y.shape[1]
```

```python
#Input Layer

model=Sequential()
model.add(Dense(100,input_shape=(30,)))
model.add(Activation('relu'))
model.add(Dropout(0.2))
# 2nd Layer

model.add(Dense(200))
model.add(Activation('relu'))
model.add(Dropout(0.2))
#3rd Layer

model.add(Dense(100))
model.add(Activation('relu'))
model.add(Dropout(0.2))
#Output Layer

model.add(Dense(num_lables))
model.add(Activation('softmax'))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
from tensorflow.keras.callbacks import ModelCheckpoint
from datetime import datetime

num_epochs=150
num_samples=30
checkpointer=ModelCheckpoint(filepath='/content/drive/MyDrive/datasets/saved_models/best_model.keras',verbose=1,save_best_only=True)
start=datetime.now()
model.fit(X_train,y_train,batch_size=num_samples,epochs=num_epochs,validation_data=(X_test,y_test),callbacks=[checkpointer])
duration=datetime.now()-start
print('Training completed in : ',duration)
```

```python
test_accuracy=model.evaluate(X_test,y_test,verbose=0)
print(test_accuracy[1])
```

### Testing the data

1. First u have to extract features from the input.
2. Then input those features inside the best model you have saved in your file.
3. Then by using reverse labels you have to convert it into back into the label it has predicted.

```python
FILE='/content/drive/MyDrive/datasets/soundata/engine_idling.wav'
import librosa
audio_signals,sample_rate=librosa.load(FILE,res_type='kaiser_best')
extracted_features=librosa.feature.mfcc(y=audio_signals,sr=sample_rate,n_mfcc=30)
extracted_features=np.mean(extracted_features.T,axis=0)
```

```python
print(extracted_features)
print(extracted_features.shape)
print('---------------------------------------------------')
extracted_features_reshape=extracted_features.reshape(1,-1)
print(extracted_features_reshape)
print(extracted_features_reshape.shape)
predicted_class=model.predict(extracted_features_reshape)
print(predicted_class)
predicted_class=np.argmax(predicted_class,axis=1)
print(predicted_class)
prediction=labelencoder.inverse_transform(predicted_class)
print(prediction)
```
